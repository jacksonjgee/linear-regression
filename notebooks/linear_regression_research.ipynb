## 1. Big Idea

Linear Regression is a **supervised machine-learning algorithm** used to predict a **continuous numerical value**.

It learns a relationship between:

- **input features**: the information supplied to the model
- **target variable**: the numerical value the model is trying to predict

Examples of numerical targets include:

- house prices
- salaries
- sales
- temperatures
- exam scores


## 2. When to Use Linear Regression

Linear Regression is useful when:

- the target variable is numerical
- the relationship between the inputs and target is approximately linear
- interpretability is important
- a simple baseline model is needed
- you want to understand how individual features relate to the prediction

It is often a good first model because it is fast, simple, and easy to interpret.


## 3. Simple Linear Regression

Simple Linear Regression uses **one input feature**.

$$
\hat{y} = \beta_0 + \beta_1x
$$

Where:

- $\hat{y}$ = predicted value
- $x$ = input feature
- $\beta_0$ = intercept or bias
- $\beta_1$ = coefficient or weight applied to the feature

The coefficient tells us how much the prediction changes when the input feature increases by one unit.


## 4. Multiple Linear Regression

Multiple Linear Regression uses **more than one input feature**.

$$
\hat{y} = \beta_0 + \beta_1x_1 + \beta_2x_2 + \cdots + \beta_nx_n
$$

Where:

- $x_1, x_2, \ldots, x_n$ are the input features
- $\beta_1, \beta_2, \ldots, \beta_n$ are the learned coefficients
- $\beta_0$ is the intercept
- $\hat{y}$ is the predicted value

With one feature, the model fits a line.  
With two features, it fits a plane.  
With more features, it fits a higher-dimensional **hyperplane**.


## 5. What the Model Learns

During training, Linear Regression learns:

- the intercept
- one coefficient for each feature
- the combination of coefficients that produces the smallest prediction error

A positive coefficient means that increasing the feature tends to increase the prediction.

A negative coefficient means that increasing the feature tends to decrease the prediction.

The size of a coefficient should be interpreted carefully because features may use very different scales.


## 6. Residuals and Prediction Error

A **residual** is the difference between the actual value and the predicted value.

$$
e_i = y_i - \hat{y}_i
$$

Where:

- $y_i$ = actual value
- $\hat{y}_i$ = predicted value
- $e_i$ = residual for observation $i$

A good regression model aims to make these residuals as small as possible.


## 7. Cost / Loss Function

Linear Regression is usually trained using **Ordinary Least Squares**.

The model chooses coefficients that minimise the **sum of squared residuals**:

$$
\sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

Squaring the residuals:

- prevents positive and negative errors from cancelling out
- penalises large errors more heavily
- gives the optimisation process a smooth mathematical objective

## 10. Evaluation Metrics

Regression models are usually evaluated using several metrics because each metric answers a different question.


### Mean Absolute Error — MAE

MAE is the average absolute difference between actual and predicted values.

$$
MAE = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i|
$$

Interpretation:

- easy to understand
- uses the same unit as the target
- treats all errors proportionally

Example: if MAE is 50,000, predictions are wrong by about 50,000 units on average.


### Mean Squared Error — MSE

MSE is the average squared prediction error.

$$
MSE = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2
$$

Interpretation:

- penalises large errors strongly
- useful during optimisation
- harder to interpret because the unit is squared


### Root Mean Squared Error — RMSE

RMSE is the square root of MSE.

$$
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2}
$$

Interpretation:

- uses the same unit as the target
- penalises large mistakes more heavily than MAE
- useful when large errors are especially undesirable


### R-Squared — $R^2$

$R^2$ measures how much of the variation in the target is explained by the model.

$$
R^2 = 1 - \frac{\sum (y_i - \hat{y}_i)^2}{\sum (y_i - \bar{y})^2}
$$

Typical interpretation:

- $R^2 = 1$: perfect fit
- $R^2 = 0$: no better than predicting the mean
- $R^2 < 0$: worse than predicting the mean

$R^2$ should not be used by itself. It should be interpreted alongside MAE or RMSE.


````python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
)
import numpy as np

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("RMSE:", rmse)
print("R²:", r2)
````

## 11. Assumptions of Linear Regression

Linear Regression generally works best when:

1. **Linearity**  
   The relationship between the features and target is approximately linear.

2. **Independent observations**  
   One observation should not strongly determine another.

3. **Constant error variance**  
   The spread of residuals should remain reasonably consistent.

4. **Low multicollinearity**  
   Input features should not be excessively correlated with each other.

5. **No extreme influential outliers**  
   A small number of unusual values should not dominate the fitted model.

These assumptions are mainly about whether the model is appropriate and whether its coefficients can be trusted.


## 12. Strengths

Linear Regression is:

- simple to understand
- fast to train
- easy to interpret
- useful as a baseline
- suitable for continuous numerical targets
- able to show how features influence predictions


## 13. Limitations

Linear Regression can struggle when:

- relationships are strongly non-linear
- there are extreme outliers
- important categorical variables have not been encoded
- features are highly correlated
- interactions between variables are important
- the data contains complex patterns


## 14. Feature Scaling

Linear Regression can usually be trained without feature scaling.

However, scaling may still be useful when:

- comparing coefficient sizes
- using regularised regression such as Ridge or Lasso
- combining Linear Regression with other models in a pipeline

For ordinary `LinearRegression`, scaling is not strictly required.
