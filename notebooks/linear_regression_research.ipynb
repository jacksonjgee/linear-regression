# Linear Regression Research

This notebook is used to understand the theory behind Linear Regression before applying it to the house price prediction project.

The goal is not to build the final model here. Instead, this notebook is for learning the key ideas that will help explain and improve the model later.

## What this notebook covers

- What Linear Regression is
- Why Linear Regression is useful for predicting continuous values
- The difference between features and the target variable
- How a regression line is fitted to data
- What coefficients and intercepts mean
- How model predictions are evaluated
- The meaning of MAE, MSE, RMSE, and R²
- Common limitations of Linear Regression

## Project connection

In this project, Linear Regression will be used to predict house prices.

The target variable is:

`Price`

The input features may include:

- bedrooms
- bathrooms
- living area size
- lot size
- number of floors
- property condition
- year built
- year renovated

By understanding the theory first, it will be easier to explain what the model is doing and why its predictions may or may not be accurate.
